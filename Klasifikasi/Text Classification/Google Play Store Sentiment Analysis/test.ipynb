```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
from spellchecker import SpellChecker

import re
import requests
```

```python
stop_words = set(stopwords.words('indonesian'))
stemmer = StemmerFactory().create_stemmer()

url_negatif_1 = "https://raw.githubusercontent.com/masdevid/ID-OpinionWords/master/negative.txt"
url_negatif_2 = "https://raw.githubusercontent.com/ramaprakoso/analisis-sentimen/master/kamus/negative_keyword.txt"
url_negatif_3 = "https://raw.githubusercontent.com/ramaprakoso/analisis-sentimen/master/kamus/negative_add.txt"
url_perbaikan = "https://raw.githubusercontent.com/ramaprakoso/analisis-sentimen/master/kamus/kbba.txt"


def load_words_from_url(url):
    response = requests.get(url)
    if response.status_code == 200:
        words = set(response.text.splitlines())
        return words
    else:
        print("Failed to fetch data from URL.")
        return set()


kata_negatif_1 = load_words_from_url(url_negatif_1)
kata_negatif_2 = load_words_from_url(url_negatif_2)
kata_negatif_3 = load_words_from_url(url_negatif_3)
kata_1 = kata_negatif_1.union(kata_negatif_2)

kata_perbaikan = load_words_from_url(url_perbaikan)
kata_negatif = kata_1.union(kata_negatif_3)


def jenis_kata_negatif(data, dictionary):
    penampung = []
    kalimat = preprocessing(data, dictionary)
    list_words = word_tokenize(kalimat)

    for kata in kalimat.split():
        if kata in kata_negatif:
            penampung.append(kata)

    return penampung


def perbaikan_dictionary(data):
    dictionary = {}

    for item in kata_perbaikan:
        words = item.split('\t')
        if len(words) == 2:
            dictionary[words[0]] = words[1]

    return dictionary


def preprocessing(data, dictionary):
    fixed_sentence = []
    words = re.sub(r'[^\w\s]', '', data)
    list_words = word_tokenize(str(words).lower())

    for word in list_words:
        fixed_sentence.append(dictionary.get(word, word))

    filtered_words = [word for word in fixed_sentence if word.lower() not in stop_words]
    kalimat = ' '.join(filtered_words)
    return kalimat


def stemming(data, dictionary):
    kalimat = preprocessing(data, dictionary)
    list_words = word_tokenize(kalimat)
    
    stemmed_words = [stemmer.stem(word) for word in list_words]
    kalimat = ' '.join(stemmed_words)
    kalimat = re.sub(r'\s+', ' ', kalimat)
    
    return kalimat
```

```python
def analyze_sentiment(data):
    analysis = SIA().polarity_scores(str(data['content']).lower())
    sentiment_score = TextBlob(str(data['content']).lower()).sentiment.polarity
    compound = analysis['compound']

    if (data['score'] == 3):
        if (analysis['neg'] > analysis['pos']) and sentiment_score < 0:
            sentiment = 'negatif'

        elif analysis['neg'] < analysis['pos'] and sentiment_score > 0:
            sentiment = 'positif'

        elif sentiment_score == 0 and analysis['neu'] == 1:
             sentiment = 'netral'

        else:                           
            sentiment = 'tidak terdeteksi'
    else:
        if data['score'] > 3:
           sentiment = 'positif'
           
        elif data['score'] < 3:
             sentiment = 'negatif'

    return sentiment, compound, sentiment_score
```

```python
dictionary = perbaikan_dictionary(kata_perbaikan)

kata_dict = {
    'app': 'aplikasi',
    'apk': 'aplikasi',
    'amanh': 'amanah',
    'amnah': 'amanah',
    'appnya': 'aplikasinya',
    'appny': 'aplikasinya',
    'kena': 'terkena',
    'kna': 'terkena',
    'lm': 'lama',
    'mmuaskan': 'memuaskan',
    'mksud': 'maksud',
    'oesen': 'pesan',
    'pesen': 'pesan',
    'psar': 'pasar',
    'pegi': 'pergi',
    'sllu': 'selalu',
    'sudh': 'sudah',
    'sktr': 'sekitar',  
    'sya': 'saya',
    'tdrdekat': 'terdekat'
}

dictionary.update(kata_dict)
```

```python
def analyze_sentiment(data, column):
    analysis = SIA().polarity_scores(str(data[column]).lower())
    sentiment_score = TextBlob(str(data[column]).lower()).sentiment.polarity
    compound = analysis['compound']

    if (data['score'] == 3):
        if (analysis['neg'] > analysis['pos']) and sentiment_score < 0:
            sentiment = 'negatif'

        elif analysis['neg'] < analysis['pos'] and sentiment_score > 0:
            sentiment = 'positif'

        elif sentiment_score == 0 and analysis['neu'] == 1:
             sentiment = 'netral'

        else:                           
            sentiment = 'tidak terdeteksi'
    else:
        if data['score'] > 3:
           sentiment = 'positif'
           
        elif data['score'] < 3:
             sentiment = 'negatif'

    return sentiment, compound, sentiment_score
```
